```python
import pandas as pd
from gzip import GzipFile
data = pd.read_csv(GzipFile("2013-311-Data.csv.gz"), header=0, parse_dates=[1])
```

```python
data.index = pd.DatetimeIndex(data['creation_date'])
```

```python
data['call_type'].groupby(data.index).count().resample('1W', how='sum').plot()
```

```python
data['hour'] = data.index
data['hour'] = data['hour'].apply(lambda x: x.hour)
```

```python
data['call_type'].groupby(data['hour']).count().plot()
```

```python
data.groupby(['call_type', 'hour']).count().ix['Parking Permits (CSC)']['ward'].plot()
```

```python
filtered = data.groupby(['call_type', 'hour']).count()

display = pd.DataFrame(index=data.groupby('hour').count().index)
for call_type in filtered.index.levels[0]:
    display[call_type] = filtered.ix[call_type]['ward']
```

```python
display = display.fillna(value=0)
display.plot(kind='area', legend=False)
```

```python
display.columns
```

```python
display['Recycle Bins'] = display['Retail-All Container(46L)'] + display['Retail-All Container(80L)'] + display['Retail-Black Box'] + display['Retail-Blue Box'] + display['Retail-Green 46L & KC'] + display['Retail-Green 80L & KC'] + display['Retail-Green Bin 46L'] + display['Retail-Green Bin 80L'] + display['Retail-Kitchen Container']
for name in display.columns:
    if name.startswith('Retail'):
        del display[name]
del display['Call Centre Use Only']
display['Parking'] = display['Parking Control'] + display['Parking Equipment'] + display['Parking Lots-Mun. (CSC)'] + display['Parking Permits (CSC)'] + display['Parking Ticket Inquiry']
del display['Parking Control'] 
del display['Parking Equipment'] 
del display['Parking Lots-Mun. (CSC)'] 
del display['Parking Permits (CSC)'] 
del display['Parking Ticket Inquiry']
for col in display.columns:
    if display[col].sum() < 10:
        del display[col]
```

```python
display.to_csv('hourly_data.csv')
display.to_json('hourly_data.json')
```

```python
wards_multiindex = data.groupby(['ward','call_type', 'hour']).count()
```

```python
flattened = wards_multiindex.reset_index()
```

```python
wards = {}
for ward in flattened['ward'].unique():
    wards[ward] = flattened[flattened['ward'] == ward].pivot(index='hour', columns='call_type', values='File').fillna(value=0)
    wards[ward].to_json('{0}.json'.format(ward))
```
